```python
while True:
    s = input("整数を入力（qで終了）: ")
    if s == "q":
        break
    try:
        n = int(s)
        print("偶数" if n % 2 == 0 else "奇数")
    except ValueError:
        print("整数を入力してください")
```

```python
# 解答例
subjects = ["国語", "数学", "英語"]
scores   = [80, 70, 90]

for sub, sc in zip(subjects, scores):
    print(f"{sub}: {sc}点")
```

```python
# 半答
for n in range(1, 31):
    if n % 15 == 0:
        print("FizzBuzz")
    elif n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    else:
        print(n)
```

```python
# 全答
for i in range(1,31):
    text = ""
    if i%3==0:
        text += "Fizz"
    if i%5==0:
        text += "Buzz"
    if i%3!=0 and i%5!=0:
        text = str(i)
    print(text)
        
```

```python
count_a = 0
count_b = 0
for i in range(1,101):
    if i%2==0 and i%3==0:
        count_a += 1
    if i%2==1 and i%5==0:
        count_b += 1
print(f"偶数かつ3の倍数の個数は {count_a}個，奇数かつ5の倍数の個数は {count_b}個")
```
