```python
import math
def f(x):
    return math.cos(x) - x
```

```python
def bisection(a, b, eps=1e-6):
    if f(a) * f(b) >= 0:
        print("2分法を使えません（符号が同じです）")
        return None
    count = 0
    while b - a > eps:
        m = (a + b) / 2
        if f(a) * f(m) > 0:
            a = m
        else:
            b = m
        count += 1
    print(f"反復回数{count}回")
    return (a + b) / 2
```

```python
bisection(0,1)
```

```python
import math
def f(x):
    return math.cos(x) - x

def df(x):
    return -math.sin(x) - 1
```

```python
def newton(x0, eps=1e-6, max_iter=100):
    x = x0
    for i in range(max_iter):
        x_new = x - f(x) / df(x)
        if abs(x_new - x) < eps:
            print(f"反復回数{i}回")
            return x_new
        x = x_new
    return x
```

```python
newton(2)
```
