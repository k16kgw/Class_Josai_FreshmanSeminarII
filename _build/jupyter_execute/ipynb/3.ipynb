```python
score = 85
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("D")
```

```python
s = input("整数を入力: ")# inputでキーボードからの入力値を変数に代入する
try:
    n = int(s)
    if n % 2 == 0:
        print("偶数")
    else:
        print("奇数")
except ValueError:
    print("整数を入力してください")
```
