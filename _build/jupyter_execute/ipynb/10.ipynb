```python
import random

def approximate_pi(N):
    count = 0
    for _ in range(N):
        x = random.random()
        y = random.random()
        if x*x + y*y <= 1:
            count += 1
    return 4 * count / N

print(approximate_pi(10000000))
```

```python
import math
import random

# 積分したい関数
def f(x):
    return math.sin(x)

# 正確な値
exact = 2.0

# モンテカルロ積分（平均値法）
def mc_integral(f, a, b, N):
    total = 0.0
    for _ in range(N):
        x = random.uniform(a, b)
        total += f(x)
    return (b - a) * total / N

# 区分求積法（左端）
def integral_riemann(f, a, b, n):
    width = (b - a) / n
    s = 0.0
    for i in range(n):
        x = a + i * width
        s += f(x) * width
    return s

# 台形公式
def integral_trapezoid(f, a, b, n):
    h = (b - a) / n
    s = f(a) + f(b)
    for i in range(1, n):
        x = a + i * h
        s += 2 * f(x)
    return s * h / 2

# 実験するN
Ns = [1000, 3000, 10000, 30000]

# 乱数は毎回変わるので、比較を安定させたい場合は seed を固定してもよい
random.seed(0)

# 結果を保存
results = []

for N in Ns:
    mc = mc_integral(f, 0, math.pi, N)
    riem = integral_riemann(f, 0, math.pi, N)     # 比較のため n=N とする
    trap = integral_trapezoid(f, 0, math.pi, N)   # 比較のため n=N とする

    err_mc = abs(mc - exact)
    err_riem = abs(riem - exact)
    err_trap = abs(trap - exact)

    results.append((N, mc, err_mc, riem, err_riem, trap, err_trap))

# 表示
print("N\tMC値\t\tMC誤差\t\t区分求積誤差\t台形誤差")
for N, mc, err_mc, riem, err_riem, trap, err_trap in results:
    print(f"{N}\t{mc:.10f}\t{err_mc:.3e}\t{err_riem:.3e}\t{err_trap:.3e}")

print("\n--- 誤差の倍率（モンテカルロ）---")
errs = [row[2] for row in results]  # MC誤差だけ取り出す
for i in range(1, len(Ns)):
    ratio = errs[i-1] / errs[i] if errs[i] != 0 else float("inf")
    print(f"N: {Ns[i-1]} → {Ns[i]} のとき，誤差は約 {ratio:.2f} 倍改善")
```
