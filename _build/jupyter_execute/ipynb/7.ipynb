```python
def gcd(a, b):
    while b != 0:
        r = a % b
        a, b = b, r
    return a
```

```python
def sieve(n):
    is_prime = [True] * (n+1)
    is_prime[0] = is_prime[1] = False

    for k in range(2, int(n**0.5)+1):
        if is_prime[k]:
            for x in range(k*k, n+1, k):
                is_prime[x] = False

    return [i for i in range(2, n+1) if is_prime[i]]
```

```python
def sieve(n):
    numbers = list(range(2,n+1))
    primes = []
    while len(numbers) > 0:
        p = numbers[0]
        primes.append(p)
        for i in range(1, n//p + 1):
            try:
                numbers.remove(p * i)
            except ValueError:
                pass   # すでに消されている場合は無視する
    return primes
```

```python
print(sieve(100))
```

```python
primes = sieve(100)
print(f"100以下の素数の個数は{len(primes)}個です。")
```

```python
def prime_factorize(n):
    factors = []
    k = 2

    while k*k <= n:
        while n % k == 0:
            factors.append(k)
            n //= k
        k += 1

    if n > 1:
        factors.append(n)

    return factors
```
